```python
def count(l, sub):
    '''I want to make the same function as s.count()'''
    counter = 0
    for word in l:
        if sub in word:
            counter += 1
    return counter
```

```python
count(l,'hamlet')
```

```python
dic = {}
for word in l:
    dic.setdefault(word,0)
    
for word in dic.keys():
    dic.update({word:count(l,word)})
```
