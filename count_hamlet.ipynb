```python
import re
import time

filename = './input/HAMLET.txt'
with open(filename, "rb") as f:
    s = f.read()
    
s = s.lower()
word_list = re.findall(r'[a-zA-Z]+', s)
```

#Using the .count method on the word list

```python

def word_count_A(word_list):
   word_count_dict = {}
   unique_words = list(set(word_list))
   for word in unique_words:
       cnt = word_list.count(word)
       word_count_dict[word] = cnt
   return word_count_dict
    

%timeit word_count_dict = word_count_A(word_list)
```

```python
sorted(word_count_dict.items(), key = lambda x:x[1], reverse = True)
```

###loop over the list and add 1 evertime you see the same word

```python
def word_count_B(word_list):
    word_count_dict = {}
    for word in word_list:
        word_count_dict.setdefault(word,0)
        word_count_dict[word] += 1
    return word_count_dict

%timeit word_count_dict = word_count_B(word_list)
```

```python
sorted(word_count_dict.items(), key = lambda x:x[1], reverse = True)
```
